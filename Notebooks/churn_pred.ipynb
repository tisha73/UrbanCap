```python
import pandas as pd 
import numpy as np 
import random 
```

```python
bookings = pd.read_csv("C:\\Users\\Tisha\\Downloads\\Past_Booking.csv")
```

```python
users = pd.read_excel("C:\\Users\\Tisha\\Downloads\\userUrban.xlsx")
```

```python
# Convert relevant columns to datetime
bookings['Booking_Date'] = pd.to_datetime(bookings['Booking_Date'])
users['last_booking_date'] = pd.to_datetime(users['last_booking_date'])

# Merge data
merged_df = pd.merge(users, bookings, left_on='user_id', right_on='User_ID', how='left')
```

```python
merged_df.info()
```

```python
# Group by user_id to compute aggregated features
user_features = merged_df.groupby('user_id').agg(
    Total_Bookings=('Booking_ID', 'count'),
    First_Booking_Date=('Booking_Date', 'min'),
    Last_Booking_Date=('Booking_Date', 'max'),
    Days_Since_Last_Booking=('Booking_Date', lambda x: (pd.Timestamp.today() - x.max()).days),
    Avg_Booking_Gap=('Booking_Date', lambda x: x.diff().mean().days if x.count() > 1 else np.nan),
    Service_Types_Used=('Service_ID', lambda x: list(x.unique())),
    Preferred_Service_Type=('Service_ID', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
    Total_Spend=('Final_Payment', 'sum'),
    Avg_Spend_Per_Booking=('Final_Payment', 'mean'),
    Last_Spend_Amount=('Final_Payment', 'last'),
    Discount_Usage_Frequency=('Discount_Applied', lambda x: (x > 0).sum() / x.count() if x.count() > 0 else 0),
    Cancellation_Count=('Booking_Status', lambda x: (x == 'Cancelled').sum()),
    Reschedule_Count=('Booking_Status', lambda x: (x == 'Rescheduled').sum()),
    No_Show_Count=('Booking_Status', lambda x: (x == 'No Show').sum())
).reset_index()

# High-Value Customer: If total spend > threshold (e.g., $500)
user_features['High-Value_Customer'] = (user_features['Total_Spend'] > 4000).astype(int)

# Engagement & Interaction Features (Randomized with seed)
user_features['App_Logins_Per_Month'] = np.random.randint(1, 20, size=len(user_features))
user_features['Total_App_Logins'] = user_features['App_Logins_Per_Month'] * 12
user_features['Browsing_History_Depth'] = np.random.randint(1, 10, size=len(user_features))
user_features['Search_Queries_Performed'] = np.random.randint(1, 50, size=len(user_features))
user_features['Click-Through-Rate (CTR)'] = np.random.uniform(0.01, 0.5, size=len(user_features))
user_features['Referral_Usage'] = np.random.randint(0, 5, size=len(user_features))
user_features['Social_Media_Engagement'] = np.random.randint(0, 20, size=len(user_features))
user_features['Time_Spent_on_Platform (min)'] = np.random.randint(5, 40, size=len(user_features))

# Customer Feedback & Satisfaction Features
user_features['Feedback_Score'] = users['feedback_score']
user_features['Last_Feedback_Score'] = np.random.uniform(1, 5, size=len(user_features))
user_features['Negative_Reviews_Count'] = np.random.randint(0, 10, size=len(user_features))
user_features['Complaints_Raised'] = np.random.randint(0, 5, size=len(user_features))
user_features['Issue_Resolution_Time'] = np.random.uniform(1, 10, size=len(user_features))
# user_features['Loyalty_Program_Member'] = np.random.choice([0, 1], size=len(user_features))
# user_features['Loyalty_Points_Used'] = np.random.randint(0, 100, size=len(user_features))
user_features['Customer_Support_Contacts'] = np.random.randint(0, 20, size=len(user_features))

# Behavioral Features (Churn Risk Indicators)
user_features['Inactive_Days'] = user_features['Days_Since_Last_Booking']
user_features['Booking_Frequency_Change'] = np.random.uniform(-0.5, 0.5, size=len(user_features))
user_features['Spend_Change_Rate'] = np.random.uniform(-0.5, 0.5, size=len(user_features))
user_features['Service_Type_Change'] = np.random.choice(['None', 'Beauty', 'Home Cleaning', 'Repair'], size=len(user_features))
user_features['Unsubscribed_Notifications'] = np.random.choice([0, 1], size=len(user_features))
user_features['Downgrade_Service_Usage'] = np.random.choice([0, 1], size=len(user_features))

# Churn Label (Logic-Based)
def churn_label(row):
    if row['Inactive_Days'] > 200 or row['Booking_Frequency_Change'] < -0.5 or row['Spend_Change_Rate'] < -0.5:
        return 1  # Churned
    elif row['Unsubscribed_Notifications'] == 1 and row['Downgrade_Service_Usage'] == 1:
        return 1  # Churned
    else:
        return 0  # Not churned

# Apply churn label to the DataFrame
user_features['Churn_Label'] = user_features.apply(churn_label, axis=1)
# Fill missing values
user_features.fillna(0, inplace=True)

# Merge with users DataFrame
users_final = users.merge(user_features, on='user_id', how='left')
```
